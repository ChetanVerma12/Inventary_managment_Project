# Inventory Turnover Analysis using Pandas

```

import pandas as pd

# Load data
df = pd.read_csv("inventory_turnover.csv")

df.head()
```

```

# Calculate Average Inventory and Turnover
df["Average_Inventory"] = (df["Opening_Inventory"] + df["Closing_Inventory"]) / 2
df["Turnover"] = df["COGS"] / df["Average_Inventory"]

df
```
