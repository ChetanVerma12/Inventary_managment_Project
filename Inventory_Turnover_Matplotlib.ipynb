# Inventory Turnover Visualization using Matplotlib

```python

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("inventory_turnover.csv")
```

```python

plt.plot(df["Month"], df["Turnover"], marker="o")
plt.title("Monthly Inventory Turnover")
plt.xlabel("Month")
plt.ylabel("Turnover Ratio")
plt.grid(True)
plt.show()
```
